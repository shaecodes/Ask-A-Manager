```python
df.dropna(subset=['Ethnicity'], inplace=True)
```

```python
black_descent = df.loc[df['Ethnicity'].str.contains("Black", case=False)]
asian_descent = df.loc[df['Ethnicity'].str.contains("Asian", case=False)]
mena_descent = df.loc[df['Ethnicity'].str.contains("Eastern", case=False)]
native_descent = df.loc[df['Ethnicity'].str.contains("Native", case=False)]
spanish_descent = df.loc[df['Ethnicity'].str.contains("Spanish", case=False)]
white = df.loc[df['Ethnicity'].str.contains("White", case=False)]
```

```python
black_count = len(black_descent)
asian_count = len(asian_descent)
mena_count = len(mena_descent)
native_count = len(native_descent)
spanish_count = len(spanish_descent)
white_count = len(white)

colors = sns.color_palette("pastel")

sns.barplot(x=['Black', 'Asian', 'MENA', 'Native', 'Spanish', 'White'], y=[black_count, asian_count, mena_count, native_count, spanish_count, white_count], palette=colors)
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.title('Count of Rows by Ethnicity')
plt.show()
```

## Cleaning the Data: Annual Salary

### Changing annual salary into currency 

```python
df['Annual Salary'] = df['Annual Salary'].str.replace(',', '').astype(int)
```

```python
df['Annual Salary'] = df['Annual Salary'].astype(float)
```

```python
df.head()
```

```python
index_names = df[df['Annual Salary'] == 0].index
df.drop(index_names, inplace=True)
```

### Filtering the Annual Salary for USD Currency using a percentile-based filter

```python
df_usa = df[df['Currency'] == 'USD']
percentile_lower = df_usa['Annual Salary'].quantile(0.01)
percentile_higher = df_usa['Annual Salary'].quantile(0.99)

df_filtered_usa = df_usa[(df_usa['Annual Salary'] >= percentile_lower) & (df_usa['Annual Salary'] <= percentile_higher)]
```

```python
df_filtered_usa["Annual Salary"].min()
```

```python
df_filtered_usa["Annual Salary"].max()
```
